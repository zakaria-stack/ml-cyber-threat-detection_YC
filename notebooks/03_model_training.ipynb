```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import joblib

print("⏳ Chargement des données nettoyées...")
# 1. Charger les données propres depuis la Tâche 7
X = pd.read_csv('../data/KDDTrain_cleaned.csv')
y = pd.read_csv('../data/KDDTrain_labels.csv')

# 2. Séparer en données d'entraînement (80%) et de test (20%)
# C'est comme un examen : on lui donne 80% pour réviser, et 20% pour le tester !
X_train, X_test, y_train, y_test = train_test_split(X, y.values.ravel(), test_size=0.2, random_state=42)

print("🧠 Début de l'entraînement de l'IA (Random Forest)... Veuillez patienter.")
# 3. Créer et entraîner le modèle
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)

print("✅ Entraînement terminé !")

# 4. Faire passer l'examen au modèle (Prédictions sur les 20% restants)
y_pred = rf_model.predict(X_test)

# 5. Afficher les résultats (les notes de l'examen !)
print("\n--- 📊 Résultats des performances de l'IDS ---")
print(f"Précision Globale (Accuracy) : {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nRapport Détaillé :")
print(classification_report(y_test, y_pred, target_names=['Normal (0)', 'Attaque (1)']))

# 6. Sauvegarder le cerveau de l'IA
joblib.dump(rf_model, '../models/rf_model.pkl')
print("\n💾 Le modèle a été sauvegardé dans le dossier 'models/' sous le nom 'rf_model.pkl'.")
```
